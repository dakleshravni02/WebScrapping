```python
import requests
from bs4 import BeautifulSoup

# URL of the site to scrape
url = "http://books.toscrape.com/"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content of the page
soup = BeautifulSoup(response.text, 'html.parser')

# Find all book containers
book_containers = soup.find_all('article', class_='product_pod')

# Extract titles and prices
books = []
for container in book_containers:
    title = container.h3.a['title']
    price = container.find('p', class_='price_color').text
    books.append({"title": title, "price": price})

# Print the extracted data
for book in books:
    print(f"Title: {book['title']}, Price: {book['price']}")

# Optionally, you can save the extracted data to a file or a database
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the site to scrape
url = "http://books.toscrape.com/"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content of the page
soup = BeautifulSoup(response.text, 'html.parser')

# Find all book containers
book_containers = soup.find_all('article', class_='product_pod')

# Extract titles, prices, stock status, and star ratings
books = []
for container in book_containers:
    title = container.h3.a['title']
    price = container.find('p', class_='price_color').text
    stock_status = container.find('p', class_='instock availability').text.strip()
    star_rating = container.find('p', class_='star-rating')['class'][1]  # Extract the rating class
    
    books.append({
        "title": title,
        "price": price,
        "stock_status": stock_status,
        "star_rating": star_rating
    })

# Print the extracted data in a clear format
for book in books:
    print(f"ğŸ“– Title: {book['title']}\nğŸ’² Price: {book['price']}\nğŸ“¦ Stock Status: {book['stock_status']}\nâ­ Star Rating: {book['star_rating']}\n")

# Optionally, you can save the extracted data to a file or a database
```
